# Plot initial Cpools and Optimized parameter values

```python
# import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# change font type 
##-*- coding: UTF-8 -*- 
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Times New Roman']
```

```python
# Modify default values of rcParams dictionary
parameters = {'axes.titlesize': 20,
              'axes.labelsize': 20}
plt.rcParams.update(parameters)
plt.rc('xtick', labelsize= 20) 
plt.rc('ytick', labelsize= 20) 
```

```python
parameters = {'legend.labelsize': 21}
```

## Double infer

```python
# data prepare 
model_name = 'MIMICSa'
param0 = pd.read_csv('../output/parameter_opt_' + model_name + '.csv')
```

```python
param0['fMB'] = param0['fMIC_1'] + param0['fMIC_2']
param0['fSOM_2'] = 1- param0['fMB'] - param0['fSOM_1']- param0['fSOM_3']
par0 = param0[['fSOM_1', 'fSOM_2','fSOM_3', 'fMIC_2', 'fMIC_1']]
par0.index= param0.site 
```

```python
model_name = 'MIMICSb'
param1 = pd.read_csv('../output/parameter_opt_' + model_name + '.csv')
```

```python
param1['fMB'] = param1['fMIC_1'] + param1['fMIC_2']
param1['fSOM_2'] = 1- param1['fMB'] - param1['fSOM_1']- param1['fSOM_3']
par1 = param1[['fSOM_1', 'fSOM_2','fSOM_3', 'fMIC_2', 'fMIC_1']]
par1.index= param1.site 
```

```python
model_name = 'MIMICSc'
param2 = pd.read_csv('../output/parameter_opt_' + model_name + '.csv')
```

```python
param2['fMB'] = param2['fMIC_1'] + param2['fMIC_2']
param2['fSOM_2'] = 1- param2['fMB'] - param2['fSOM_1']- param2['fSOM_3']
par2 = param2[['fSOM_1', 'fSOM_2','fSOM_3', 'fMIC_2', 'fMIC_1']]
par2.index= param2.site 
```

### Plot MIMICSa

```python
# create stacked bar chart for carbon fractions
par0.plot(kind='bar', stacked=True, 
          color=['grey', '#e5ab00','red','green', 'blue', ])
 
# labels for x & y axis
plt.xlabel('')
plt.ylabel('Carbon Fractions')
#place legend in center right of plot
plt.legend(["$y = x^2$", "$y = x$"], loc='upper left')

plt.xticks(rotation=45, ha='center')
plt.rcParams['legend.fontsize'] = 20

plt.legend(["$C_{Phs}$", "$C_{Chm}$", "$C_{Avl}$",  "$MB_K$", "$MB_r$"   ],
           bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.savefig('Fig.2-a.jpg', dpi=300, quality=80, 
            optimize=True, progressive=True, bbox_inches="tight") 
```

```python

plt.axhline(y = 1, color = 'gray', linestyle = '-')
plt.scatter(param0.site, param0.beta)
plt.ylim(0.8, 1.4)
plt.xticks(rotation=45, ha='center')
# specifying horizontal line type


plt.ylabel(r"$\beta$")
plt.savefig('Fig.2-d.jpg', dpi=300, quality=80, 
            optimize=True, progressive=True, bbox_inches="tight") 
```
